# Bruno Borguesan - Walking Detector.

OBJETIVO:
    - Utilizar visão computacional para identificar no video 'people-walking.mp4' pessoas falando, gerando como saída um resultado similar ao vídeo 'people-walking-processed.avi'.

PROCESSO REALIZADO:
    - Utilizei as bibliotecas openCV[0] e o método de Histogram of Oriented Gradients (HOG)[0] crianto um vetor de caracteristicas que foi aplicado ao algoritmo de Support Vector Machine (SVM) para identificação de objetos;
    - Para evitar a fragmentação de objetos utilizamos o método Non-Maximum Suppression do imutils[1];
    - Resultados não foram satistatorios mesmo testandos varios parametros diferentes (winStride, padding, scale), devido a simplicidade do algoritmo;
    - Para contornar o problema utilizei a YOLOv3[2], que usa uma abordagem de Deep Learning que separa o frame em um grid, onde cada celula vai ter um vetor associado retornando se tem um objeto/pessoa, o que é e a localização no frame. (Foi adaptado para mostrar apenas pessoas);
    - Utilizando a versão do modelo 'yolov3.weights' os resultados foram bem satisfatorios, porem, necessita de um poder computacional alto para conseguir fazer a analise 'ao vivo';
    

PROBLEMAS A SEREM RESOLVIDOS:
    - alguns frames perdem a identificação das pessoas;

SOLUÇÕES POSSÍVEIS:
    - Aplicar alguma abordagem que utilize a média de frames;

OBS:
    - YOLO também permite criar a detecção de um objeto, desde que treinado com uma quantia de dados anotados. Pode ser interessante para abordagem de interpretação de dados de vigilancia de esteira de minerios[3];

```python
import imutils
from imutils.object_detection import non_max_suppression
import cv2
import numpy as np

hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

video_capture = cv2.VideoCapture("people-walking.mp4") #Webcam object
frame_width = 640
frame_height = 360
#out = cv2.VideoWriter('outpy.avi',cv2.VideoWriter_fourcc('M','J','P','G'), 25, (frame_width,frame_height))

while True:
    ret, frame = video_capture.read()
    if not ret: break
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    #img, hitThreshold=None, winStride=None, padding=None, scale=None, finalThreshold=None, useMeanshiftGrouping=None, /
    (rects, weights) = hog.detectMultiScale(frame, winStride=(8, 8), padding=(24, 24), scale=1.05,)
    rects = np.array([[x, y, x + w, y + h] for (x, y, w, h) in rects])
    pick = non_max_suppression(rects, probs=None, overlapThresh=0.85)
    person = 1
    for (xA, yA, xB, yB) in pick:
        cv2.rectangle(frame, (xA, yA), (xB, yB), (0, 255, 0), 2)
        person += 1
    cv2.putText(frame, f'Total Persons : {person-1}', (40,70), cv2.FONT_HERSHEY_DUPLEX, 0.8, (255,0,0), 2)
    #cv2.imshow("output", frame)
    #out.write(frame)
    if cv2.waitKey(1) & 0xFF == ord('q'): #Exit program when the user presses 'q'
        break

video_capture.release()
#out.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

import time

CONFIDENCE = 0.5
SCORE_THRESHOLD = 0.5
IOU_THRESHOLD = 0.5
config_path = "yolov3-tiny.cfg" # 5~10X mais rapido, mas não tão preciso
weights_path = "yolov3-tiny.weights"
config_path = "yolov3.cfg"
weights_path = "yolov3.weights"
font_scale = 1
thickness = 1
LABELS = open("coco.names").read().strip().split("\n")
COLORS = np.random.randint(0, 255, size=(len(LABELS), 3), dtype="uint8")

net = cv2.dnn.readNetFromDarknet(config_path, weights_path)
net.setPreferableBackend(cv2.dnn.DNN_BACKEND_OPENCV)
# net.setPreferableTarget(cv2.dnn.DNN_TARGET_CPU)

ln = net.getLayerNames()
ln = [ln[i[0] - 1] for i in net.getUnconnectedOutLayers()]

#cap = cv2.VideoCapture(0)
#cap = cv2.VideoCapture("walking/view-GL2.mp4") #Webcam object
cap = cv2.VideoCapture("people-walking.mp4") #Webcam object
frame_width = 640
frame_height = 360
#out = cv2.VideoWriter('outpy2.avi',cv2.VideoWriter_fourcc('M','J','P','G'), 25, (frame_width,frame_height))

while True:
    ret, image = cap.read()
    if not ret: break
    h, w = image.shape[:2]
    blob = cv2.dnn.blobFromImage(image, 1/255.0, (416, 416), swapRB=True, crop=False)
    net.setInput(blob)
    start = time.perf_counter()
    layer_outputs = net.forward(ln)
    time_took = time.perf_counter() - start
    print("Time took:", time_took)
    boxes, confidences, class_ids = [], [], []

    # loop over each of the layer outputs
    for output in layer_outputs:
        # loop over each of the object detections
        for detection in output:
            # extract the class id (label) and confidence (as a probability) of
            # the current object detection
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            # discard weak predictions by ensuring the detected
            # probability is greater than the minimum probability
            if confidence > CONFIDENCE:
                # scale the bounding box coordinates back relative to the
                # size of the image, keeping in mind that YOLO actually
                # returns the center (x, y)-coordinates of the bounding
                # box followed by the boxes' width and height
                box = detection[:4] * np.array([w, h, w, h])
                (centerX, centerY, width, height) = box.astype("int")

                # use the center (x, y)-coordinates to derive the top and
                # and left corner of the bounding box
                x = int(centerX - (width / 2))
                y = int(centerY - (height / 2))

                # update our list of bounding box coordinates, confidences,
                # and class IDs
                boxes.append([x, y, int(width), int(height)])
                confidences.append(float(confidence))
                class_ids.append(class_id)

    # perform the non maximum suppression given the scores defined before
    idxs = cv2.dnn.NMSBoxes(boxes, confidences, SCORE_THRESHOLD, IOU_THRESHOLD)

    font_scale = 1
    thickness = 1
    person = 1
    # ensure at least one detection exists
    if len(idxs) > 0:
        # loop over the indexes we are keeping        
        for i in idxs.flatten():
            if class_ids[i] != 0: continue
            person += 1
            # extract the bounding box coordinates
            x, y = boxes[i][0], boxes[i][1]
            w, h = boxes[i][2], boxes[i][3]
            # draw a bounding box rectangle and label on the image
            cv2.rectangle(image, (x, y), (x + w, y + h), (0,255,255), thickness=2)
    cv2.putText(image, f'Total Persons : {person-1}', (40,70), cv2.FONT_HERSHEY_DUPLEX, 0.8, (255,0,0), 2)
    #cv2.imshow("image", image)
    #out.write(image)
    if cv2.waitKey(1) & 0xFF == ord('q'): #Exit program when the user presses 'q'
        break

cap.release()
#out.release()
cv2.destroyAllWindows()
```
